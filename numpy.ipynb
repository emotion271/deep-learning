### 30.计算 softmax 分数

```python
# 计算 sepallength 的 softmax 分数
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = np.genfromtxt(url, delimiter=',', dtype='object')
sepallength = np.array([float(row[0]) for row in iris])
sepallength
```

### 40.将一个数值转换为一个类别（文本）数组

```python
# 将 iris_2d 的 petallength（第三列）转换以构建一个文本数组，按如下规则进行转换:
# Less than 3 –> ‘small’
# 3-5 –> 'medium'
# '>=5 –> 'large'
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = np.genfromtxt(url, delimiter=',', dtype='object')
names = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')
```

```python
petal_length_bin = np.digitize(iris[:, 2].astype('float'), [0, 3, 5, 10])
petal_length_bin
```

digitize(data,bins)函数执行的过程大概如下：相当于两层for循环。首先最外面的一层：顺序的遍历data列表，取出当前的一个数a，然后顺序的遍历bins列表，返回bins列表中某个数的索引，如何找呢？ 规则如下：如果：bins列表是升序的，那么如果满足 bins[i-1]<=a<bins[i],那么就保存i，然后回到外层循环继续上面的操作，如果bins列表是降序的，那么如果满足 bins[i]<=a<bins[i-1],那么就保存i，当data列表遍历完之后，就返回由i组成的来源于bins的索引的一个索引列表。

```python
label_map = {1: 'small', 2: 'medium', 3: 'large', 4: np.nan}
petal_length_cat = [label_map[x] for x in petal_length_bin]
petal_length_cat[:5]
```

### 41.基于 NumPy 数组现有列创建一个新的列

```python
# 为 iris_2d 中的 volume 列创建一个新的列，volume 指 (pi x petallength x sepal_length^2)/3
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='object')
names = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')
```

```python
sepallength = iris_2d[:, 0].astype('float')
petallength = iris_2d[:, 0].astype('float')
volume = (np.pi * petallength *(sepallength**2))/3
volume.shape
```

```python
volume = volume[:, np.newaxis]
# np.newaxis 在使用和功能上等价于 None，查看源码发现：newaxis = None，其实就是 None 的一个别名。
volume.shape
```

### 67.计算 NumPy 数组的移动平均数

```python
# 计算 window size 为 3 的移动平均数
np.random.seed(100)
Z = np.random.randint(10, size=10)
Z
```

```python
# method 1
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
moving_average(Z, n=3).round(2)
```

```python
# method 2
np.convolve(Z, np.ones(3)/3, mode='valid')
```

### 70.基于给定的 1 维数组创建 strides

```python
# 给定 1 维数组 arr，使用 strides 生成一个 2 维矩阵，其中 window length 等于 4，strides 等于 2
# 例如 [[0,1,2,3], [2,3,4,5], [4,5,6,7]..]
def gen_strides(a, stride_len=5, window_len=5):
    n_strides = ((a.size-window_len)//stride_len) + 1
    return np.array([a[s:(s+window_len)] for s in np.arange(0, n_strides*stride_len, stride_len)])
```

```python
print(gen_strides(np.arange(15), stride_len=2, window_len=4))
```
